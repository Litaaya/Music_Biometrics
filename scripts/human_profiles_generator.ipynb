```python
import os
import uuid
import pandas as pd
import numpy as np
from datetime import datetime, date
from google.colab import drive

drive.mount('/content/drive')

print("Initializing User Profile Generator.")

raw_users = [
    {
        "username": "quan_do_minh",
        "dob": "2002-10-28",
        "gender": "M",
        "cultural_region": "vi",
        "baseline_resting_hr": 65.4,
        "baseline_hrv_sdnn": 55.2
    },
    {
        "username": "cece_faith",
        "dob": "1998-10-22",
        "gender": "F",
        "cultural_region": "us",
        "baseline_resting_hr": 72.1,
        "baseline_hrv_sdnn": 48.5
    },
    {
        "username": "minjun_kim",
        "dob": "2001-03-05",
        "gender": "M",
        "cultural_region": "kr",
        "baseline_resting_hr": 61.8,
        "baseline_hrv_sdnn": 62.0
    },
    {
        "username": "sakura_sato",
        "dob": "2003-12-12",
        "gender": "M",
        "cultural_region": "jp",
        "baseline_resting_hr": 74.5,
        "baseline_hrv_sdnn": 42.1
    },
    {
        "username": "sage_morgan_burke",
        "dob": "1995-07-19",
        "gender": "F",
        "cultural_region": "us",
        "baseline_resting_hr": 68.0,
        "baseline_hrv_sdnn": 50.0
    }
]

processed_records = []
current_time = datetime.now()

for user in raw_users:
    user_id = str(uuid.uuid4())

    record = {
        "user_id": user_id,
        "username": user["username"],
        "dob": datetime.strptime(user["dob"], "%Y-%m-%d").date(),
        "gender": user["gender"],
        "cultural_region": user["cultural_region"],
        "baseline_resting_hr": float(user["baseline_resting_hr"]),
        "baseline_hrv_sdnn": float(user["baseline_hrv_sdnn"]),
        "created_at": current_time,
        "updated_at": current_time
    }
    processed_records.append(record)

df_users = pd.DataFrame(processed_records)

schema_mapping = {
    'user_id': 'string',
    'username': 'string',
    'gender': 'string',
    'cultural_region': 'string',
    'baseline_resting_hr': 'float64',
    'baseline_hrv_sdnn': 'float64'
}
df_users = df_users.astype(schema_mapping)

df_users['dob'] = pd.to_datetime(df_users['dob']).dt.date

OUTPUT_PATH = "/content/drive/MyDrive/user_profiles.parquet"
df_users.to_parquet(OUTPUT_PATH, index=False, engine='pyarrow')
```
